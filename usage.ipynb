# 6.1.Data Load & Preprocess

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset using utils.load()
import utils

tracks = utils.load('data/fma_metadata/tracks.csv')
genres = utils.load('data/fma_metadata/genres.csv')
features = utils.load('data/fma_metadata/features.csv')
echonest = utils.load('data/fma_metadata/echonest.csv')

# Check the shapes of the loaded datasets
print("Tracks:", tracks.shape)
print("Genres:", genres.shape)
print("Features:", features.shape)
print("Echonest:", echonest.shape)

# Extracting MFCC features for training
small = tracks['set', 'subset'] <= 'small'
train = tracks['set', 'split'] == 'training'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[small & train, ('track', 'genre_top')]
y_test = tracks.loc[small & test, ('track', 'genre_top')]
X_train = features.loc[small & train, 'mfcc']
X_test = features.loc[small & test, 'mfcc']

print(f"{y_train.size} training examples, {y_test.size} testing examples")
print(f"{X_train.shape[1]} features, {np.unique(y_train).size} classes")

# â Fix 1: Scale Features to Improve Convergence
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# â Fix 2: Check for Missing Values and Handle Them
print("Missing values in X_train:", np.isnan(X_train).sum())
print("Missing values in X_test:", np.isnan(X_test).sum())

X_train = np.nan_to_num(X_train)
X_test = np.nan_to_num(X_test)

# â Fix 3: Reduce Feature Dimensions Using PCA (If Needed)
if X_train.shape[1] > 200:  # Apply PCA if too many features
    print("Applying PCA to reduce features...")
    pca = PCA(n_components=50)
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)

# â Fix 4: Train Logistic Regression with Improved Settings
lr_model = LogisticRegression(max_iter=2000, solver='saga', random_state=42)
lr_model.fit(X_train, y_train)

# Predict on test set
y_pred_lr = lr_model.predict(X_test)

# Evaluate performance
lr_accuracy = accuracy_score(y_test, y_pred_lr)
lr_f1 = f1_score(y_test, y_pred_lr, average='macro')
lr_confusion = confusion_matrix(y_test, y_pred_lr)

print("Logistic Regression Accuracy:", lr_accuracy)
print("Logistic Regression F1-score:", lr_f1)
print("Logistic Regression Confusion Matrix:\n", lr_confusion)
```

# k-Nearest Neighbors (kNN) with Optimized Learning

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder

# Encode genre labels as numerical values
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Initialize kNN model
knn_model = KNeighborsClassifier()

# Define hyperparameter search space
param_grid = {'n_neighbors': [3, 5, 7, 9, 11],  # Different k values
              'weights': ['uniform', 'distance'],  # Weighting strategies
              'metric': ['euclidean', 'manhattan']}  # Distance metrics

# Perform Grid Search to optimize kNN
grid_search = GridSearchCV(knn_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train_encoded)  # Use encoded labels

# Get the best kNN model
best_knn = grid_search.best_estimator_

# Predict on test set
y_pred_knn = best_knn.predict(X_test)

# Decode predicted values back to original genre labels (optional)
y_pred_knn_labels = label_encoder.inverse_transform(y_pred_knn)

# Evaluate performance
knn_accuracy = accuracy_score(y_test_encoded, y_pred_knn)
knn_f1 = f1_score(y_test_encoded, y_pred_knn, average='macro')
knn_confusion = confusion_matrix(y_test_encoded, y_pred_knn)

print("â Best kNN Parameters:", grid_search.best_params_)
print("kNN Accuracy:", knn_accuracy)
print("kNN F1-score:", knn_f1)
print("kNN Confusion Matrix:\n", knn_confusion)
```

# step1: Load Data

```python
import numpy as np
import pandas as pd
import utils  # Using your utils.py
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.model_selection import GridSearchCV

# ð Load data using your utils.py
print("\nð Loading Data...")
tracks = utils.load('data/fma_metadata/tracks.csv')
features = utils.load('data/fma_metadata/features.csv')
genres = utils.load('data/fma_metadata/genres.csv')

# â Map genre_id to genre names
genre_mapping = genres['title'].to_dict()
tracks[('track', 'genre_top')] = tracks[('track', 'genre_top')].map(genre_mapping)

# â Check for unmapped genres
if tracks[('track', 'genre_top')].isna().sum() > 0:
    print("\nâ ï¸ Warning: Some genres were not mapped correctly!")

# ð Print dataset shapes
print(f"â Tracks: {tracks.shape}, Features: {features.shape}, Genres: {genres.shape}")
```

# Step 2: Select Features

```python
# ð Selecting Relevant Features (Matching MultiIndex Structure)
selected_features = [
    ('chroma_cens', 'mean', '01'), ('chroma_cens', 'std', '01'),  # Chroma (12-dim)
    ('tonnetz', 'mean', '01'), ('tonnetz', 'std', '01'),          # Tonnetz (6-dim)
    ('mfcc', 'mean', '01'), ('mfcc', 'std', '01'),                # MFCC (13-dim)
    ('spectral_centroid', 'mean', '01'), ('spectral_centroid', 'std', '01'),
    ('spectral_bandwidth', 'mean', '01'), ('spectral_bandwidth', 'std', '01'),
    ('spectral_contrast', 'mean', '01'), ('spectral_contrast', 'std', '01'),  # Spectral Contrast (6-dim)
    ('spectral_rolloff', 'mean', '01'), ('spectral_rolloff', 'std', '01'),
    ('rmse', 'mean', '01'), ('rmse', 'std', '01'),  # RMS Energy
    ('zcr', 'mean', '01'), ('zcr', 'std', '01')     # Zero-Crossing Rate
]

# Ensure features exist in DataFrame
existing_features = [col for col in selected_features if col in features.columns]

# Extract X (features) and y (labels)
X = features[existing_features].values
y = tracks[('track', 'genre_top')].values  # Genre names

# â Print shape
print(f"â Selected Features Shape: {X.shape}, Labels: {y.shape}")
```

# Step 3: Preprocess Data & Split

```python
from sklearn.preprocessing import LabelEncoder

# ð¯ Encode genre labels into numbers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# ð Train-test split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# ð¬ Standardize Features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(f"â Data Ready: {X_train.shape} training samples, {X_test.shape} testing samples")
```
